```python
import pandas as pd
import numpy as np
from time import time
import datetime
from datetime import date
from datetime import datetime
import collections
from collections import defaultdict
from pandas import Series, DataFrame
import datetime as dt

# for warnings 
import warnings
warnings.filterwarnings('ignore')
```

```python
enquiry = pd.read_csv('raw_enquiry_30_new.csv')
data = pd.read_csv('raw_data_30_new.csv')
account = pd.read_csv('raw_account_30_new.csv')
```

# Creating the New Features 

## count_enquiry_recency_365

```python
enquiry_dt = pd.to_datetime(enquiry['enquiry_dt'])
dt_opened = pd.to_datetime(enquiry['dt_opened'])
```

```python
enquiry['count_enquiry_recency_365'] = dt_opened - enquiry_dt 
enquiry.head()
```

```python
f = enquiry[(enquiry.count_enquiry_recency_365<= '365days')]
group = f.groupby(f['customer_no'])
group1 = group.count()
enqCount = group1.loc[:,['count_enquiry_recency_365']]
data_1 = data.join(enqCount, on='customer_no')
data_1.count_enquiry_recency_365.fillna(0, inplace=True)
data_1.head()
```

## count_enquiry_recency_90

```python
enquiry['count_enquiry_recency_90'] = dt_opened - enquiry_dt 
enquiry.head()
```

```python
ff = enquiry[(enquiry.count_enquiry_recency_90<= '90days')]
gg = ff.groupby(ff['customer_no'])
hh = gg.count()
enqCount_ninety = hh.loc[:,['count_enquiry_recency_90']]
data_2 = data_1.join(enqCount_ninety, on='customer_no')
data_2.count_enquiry_recency_90.fillna(0, inplace=True)
data_2.head()
```

## mean_diff_open_enquiry_dt

```python
enquiry['mean_diff_open_enquiry_dt'] = dt_opened - enquiry_dt 
enquiry.head()
```

```python
lll = enquiry.groupby("customer_no")["mean_diff_open_enquiry_dt"].apply(lambda x: x.mean())
llll= lll.dt.days
```

```python
data_3 = data_2.join(llll, on='customer_no')
data_3.mean_diff_open_enquiry_dt.fillna(0, inplace=True)
data_3.head()
```

## max_freq_enquiry

```python
ppp = enquiry.groupby("customer_no")["enq_purpose"].apply(lambda x: x.max()) # To get most frequent enquiry purpose
data_4 = data_3.join(ppp, on='customer_no')
data_4.rename(columns={ 'enq_purpose' : 'max_freq_enquiry' },inplace=True) # rename the lable to max_freq_enquiry
```

```python
data_4["max_freq_enquiry"] = data_4["max_freq_enquiry"].fillna(0.0).astype(int)
data_4["count_enquiry_recency_365"] = data_4["count_enquiry_recency_365"].fillna(0.0).astype(int)
data_4["count_enquiry_recency_90"] = data_4["count_enquiry_recency_90"].fillna(0.0).astype(int)
data_4["mean_diff_open_enquiry_dt"] = data_4["mean_diff_open_enquiry_dt"].fillna(0.0).astype(int)
```

```python
data_4.head()
```

## mean_diff_lastpaymt_opened_dt

```python
account.head()
```

```python
account.shape
```

```python
account = account.sort_values('customer_no',ascending=True)
```

```python
last_paymt_dt = pd.to_datetime(account['last_paymt_dt'])
opened_dt = pd.to_datetime(account['opened_dt'])
```

```python
account['mean_diff_lastpaymt_opened_dt'] = last_paymt_dt - opened_dt 
```

```python
m = account.groupby("customer_no")["mean_diff_lastpaymt_opened_dt"].apply(lambda x: x.mean())
mm= m.dt.days
data_5 = data_4.join(mm, on='customer_no')
data_5.mean_diff_lastpaymt_opened_dt.fillna(0, inplace=True)
data_5["mean_diff_lastpaymt_opened_dt"] = data_5["mean_diff_lastpaymt_opened_dt"].fillna(0.0).astype(int)
data_5.head()
```

## Ratio_currbalance_creditlimit 
### [total cur_bal_amt / total credit limit]

```python
total_cur_bal_amt = account.groupby("customer_no")["cur_balance_amt"].apply(lambda x: x.sum())
```

```python
total_credit_limit = account.groupby("customer_no")["creditlimit"].apply(lambda x: x.sum())
```

```python
account['Ratio_currbalance_creditlimit'] = total_cur_bal_amt / total_credit_limit
```

```python
data_6 = data_5.join(account['Ratio_currbalance_creditlimit'], on='customer_no')
data_6.Ratio_currbalance_creditlimit.fillna(0, inplace=True)
```

```python
data_6.Ratio_currbalance_creditlimit[np.isneginf(data_6.Ratio_currbalance_creditlimit)] = 0
data_6.Ratio_currbalance_creditlimit[np.isinf(data_6.Ratio_currbalance_creditlimit)] = 0
data_6["Ratio_currbalance_creditlimit"] = data_6["Ratio_currbalance_creditlimit"].fillna(0.0).astype(int)
```

```python
data_6.head()
```

## utilisation_trend 
### total cur_bal_amt / total creditlimit]/[mean cur_bal_amt / (meancredit limit+ mean_cashlimit)]

```python
mean_cur_bal_amt = account.groupby("customer_no")["cur_balance_amt"].apply(lambda x: x.mean())
mean_credit_limit = account.groupby("customer_no")["creditlimit"].apply(lambda x: x.mean())
mean_cashlimit = account.groupby("customer_no")["cashlimit"].apply(lambda x: x.mean())
```

```python
account['utilisation_trend'] = data_6.Ratio_currbalance_creditlimit / (mean_cur_bal_amt/(mean_credit_limit+mean_cashlimit))
```

```python
data_7 = data_6.join(account['utilisation_trend'], on='customer_no')
data_7.utilisation_trend.fillna(0, inplace=True)
data_7.utilisation_trend[np.isneginf(data_7.utilisation_trend)] = 0
data_7.utilisation_trend[np.isinf(data_7.utilisation_trend)] = 0
data_7["utilisation_trend"] = data_7["utilisation_trend"].fillna(0.0).astype(int)
```

```python
data_7.head()
```

## payment_history_mean_length

```python
account['payment_history_mean_length'] = account.paymenthistory1.apply(lambda x: len(x))
```

```python
account['payment_history_mean_length'] = account.groupby("customer_no")["payment_history_mean_length"].apply(lambda x: x.mean())
```

```python
data_8 = data_7.join(account['payment_history_mean_length'], on='customer_no')
data_8.payment_history_mean_length.fillna(0, inplace=True)
data_8.payment_history_mean_length.round()
data_8["payment_history_mean_length"] = data_8["payment_history_mean_length"].astype(int)
```

```python
data_8.head()
```

## The averge enquiry amount 

```python
account['avg_enq_amount'] = enquiry.groupby("customer_no")["enq_amt"].apply(lambda x: x.mean())
```

```python
data_9 = data_8.join(account['avg_enq_amount'], on='customer_no')
data_9.avg_enq_amount.fillna(0, inplace=True)
data_9["avg_enq_amount"] = data_9["avg_enq_amount"].fillna(0.0).astype(int)
data_9.head()
```

## max_diff_payment_start_end_dt

```python
paymt_str_dt= pd.to_datetime(account['paymt_str_dt'])
paymt_end_dt = pd.to_datetime(account['paymt_end_dt'])
```
